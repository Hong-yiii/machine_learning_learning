```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt # for making figures
%matplotlib inline
```

```python
# read in all the words
words = open('names.txt', 'r').read().splitlines()
words[:8]
```

```python
len(words)
```

```python
# build the vocabulary of characters and mappings to/from integers
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
vocab_size = len(itos)
print(itos)
print(vocab_size)
```

```python
# build the dataset
block_size = 3 # context length: how many characters do we take to predict the next one?

def build_dataset(words):  
  X, Y = [], []
  
  for w in words:
    context = [0] * block_size
    for ch in w + '.':
      ix = stoi[ch]
      X.append(context)
      Y.append(ix)
      context = context[1:] + [ix] # crop and append
  
  X = torch.tensor(X)
  Y = torch.tensor(Y)
  print(X.shape, Y.shape)
  return X, Y

import random
random.seed(42)
random.shuffle(words)
n1 = int(0.8*len(words))
n2 = int(0.9*len(words))

Xtr,  Ytr  = build_dataset(words[:n1])     # 80%
Xdev, Ydev = build_dataset(words[n1:n2])   # 10%
Xte,  Yte  = build_dataset(words[n2:])     # 10%
```

```python
# MLP revisited
n_embd = 10 # the dimensionality of the character embedding vectors
n_hidden = 200 # the number of neurons in the hidden layer of the MLP

g = torch.Generator().manual_seed(2147483647) # for reproducibility
C  = torch.randn((vocab_size, n_embd),            generator=g)
W1 = torch.randn((n_embd * block_size, n_hidden), generator=g) * (5/3)/((n_embd * block_size)**0.5) #* 0.2
#b1 = torch.randn(n_hidden,                        generator=g) * 0.01
W2 = torch.randn((n_hidden, vocab_size),          generator=g) * 0.01
b2 = torch.randn(vocab_size,                      generator=g) * 0

# BatchNorm parameters
bngain = torch.ones((1, n_hidden))
bnbias = torch.zeros((1, n_hidden))
bnmean_running = torch.zeros((1, n_hidden))
bnstd_running = torch.ones((1, n_hidden))

parameters = [C, W1, W2, b2, bngain, bnbias]
print(sum(p.nelement() for p in parameters)) # number of parameters in total
for p in parameters:
  p.requires_grad = True
```

# Batch normalization, gain and bias






#machine-learning 
*** 
Batch normalization (BN) is a powerful technique to stabilize and accelerate the training of deep neural networks. The mathematical intuition behind introducing **gain** ($\text{bngain}$) and **bias** ($\text{bnbias}$) in batch normalization lies in the balance between normalization and the network's capacity to learn rich representations.

Let’s break this down step by step:

---

### 1. **What Does Batch Normalization Do?**
Batch normalization normalizes the pre-activations of each neuron in a layer across a mini-batch of data:
$$
\hat{h}_{\text{preact}} = \frac{h_{\text{preact}} - \mu}{\sigma}
$$
where:
- $h_{\text{preact}}$ is the pre-activation output from the layer.
- $\mu$ is the mean of $h_{\text{preact}}$ across the mini-batch.
- $\sigma$ is the standard deviation of $h_{\text{preact}}$ across the mini-batch.

This normalization ensures that activations have zero mean and unit variance, which mitigates issues like exploding/vanishing gradients and helps the optimization process.

---

### 2. **Why Introduce Gain ($\text{bngain}$) and Bias ($\text{bnbias}$)?**
If we only normalize to zero mean and unit variance, we constrain the outputs of the layer to a very specific distribution. While this makes training stable, it limits the representational power of the network.

To restore flexibility, **gain** ($\text{bngain}$) and **bias** ($\text{bnbias}$) are introduced:
$$
h_{\text{preact\_normalized}} = \text{bngain} \cdot \hat{h}_{\text{preact}} + \text{bnbias}
$$

- $\text{bngain}$: Controls the scale (amplitude) of the activations.
- $\text{bnbias}$: Controls the offset (translation) of the activations.

This allows the network to learn the optimal distribution of activations during training, rather than being fixed to zero mean and unit variance.

---

### 3. **How Do Gain and Bias Differ from Weights and Biases in Activation Layers?**
While both the gain and bias in batch normalization and the weights and biases in activation layers contribute to the transformations of activations, they serve distinct roles:

1. **Weights and Biases in Layers**:
   - Transform inputs to outputs (e.g., $h = W \cdot x + b$).
   - Their primary role is to enable the network to learn mappings between inputs and desired outputs.

2. **Gain and Bias in Batch Normalization**:
   - Operate on normalized activations to restore the network’s representational capacity.
   - Their role is to adjust the scale and shift of the normalized activations to better fit the data distribution or task.

---

### 4. **Why Are Gain and Bias Optimized During Training?**
The optimization of $\text{bngain}$ and $\text{bnbias}$ allows the network to learn how much normalization is beneficial for each neuron. If these parameters were fixed:
- Some neurons might not learn efficiently because they require different distributions of activations to capture specific features.
- Batch normalization would rigidly enforce zero mean and unit variance, potentially impairing the flexibility of the network.

By optimizing these parameters, the network finds the balance between the benefits of normalization (stability, faster training) and the flexibility of learning diverse feature distributions.

---

### 5. **Mathematical Intuition**
After normalization, the transformed pre-activation for a single neuron becomes:
$$
h_{\text{post-BN}} = \text{bngain} \cdot \frac{h_{\text{preact}} - \mu}{\sigma} + \text{bnbias}
$$

- The normalization term ($\frac{h_{\text{preact}} - \mu}{\sigma}$) ensures stable gradients and prevents dead neurons.
- The gain ($\text{bngain}$) rescales the normalized values, allowing the network to learn features with varying magnitudes.
- The bias ($\text{bnbias}$) shifts the distribution, enabling the network to learn features that depend on specific offsets.

Thus, $\text{bngain}$ and $\text{bnbias}$ function as fine-tuning parameters that adapt the normalized distribution to suit the network’s requirements.

---

### 6. **Running Mean and Variance**
The lines with `bnmean_running` and `bnstd_running` in your code maintain a running estimate of the mean and standard deviation for use during inference (where batch statistics are unavailable). This ensures consistent behavior between training and testing phases.

---

### Summary
Batch normalization’s gain and bias parameters are optimized to provide the network with enough flexibility while still reaping the benefits of normalization. They function as adjustments to the normalized values, enabling the network to learn richer and more diverse features while avoiding dead neurons and training instability.

The optimization of the batch normalization (BN) parameters — **batch gain** ($\text{bngain}$) and **batch bias** ($\text{bnbias}$) — happens as part of the standard backpropagation process. They are trained just like other parameters in the network (e.g., weights and biases of layers), but their specific role is distinct. Let’s walk through the mechanics:

---

### 1. **Role of Batch Gain and Bias**
- **Batch gain ($\text{bngain}$)** and **batch bias ($\text{bnbias}$)** are trainable parameters introduced to restore the representational flexibility of the network after normalization.
- They adapt the normalized activations to the desired range or scale suitable for the task.

### 2. **How They Are Optimized**
#### Backpropagation and the Chain Rule
In backpropagation:
1. Compute the gradients of the loss $L$ with respect to the trainable parameters (weights, biases, $\text{bngain}$, and $\text{bnbias}$).
2. Update the parameters using an optimization algorithm (e.g., SGD, Adam).

For batch normalization:
- $\text{bngain}$ and $\text{bnbias}$ are optimized directly with respect to the **loss function** $L$, just like the weights in other layers.
- The gradients of the loss with respect to $\text{bngain}$ and $\text{bnbias}$ are calculated using the chain rule.

#### Example:
For a single neuron, the post-BN output is:
$$
h_{\text{post-BN}} = \text{bngain} \cdot \hat{h}_{\text{preact}} + \text{bnbias}
$$
where $\hat{h}_{\text{preact}} = \frac{h_{\text{preact}} - \mu}{\sigma}$.

- During backprop, the loss $L$ is differentiated with respect to $\text{bngain}$ and $\text{bnbias}$:
  $$
  \frac{\partial L}{\partial \text{bngain}} = \sum_i \frac{\partial L}{\partial h_{\text{post-BN}}} \cdot \hat{h}_{\text{preact}}
$$
  $$
  \frac{\partial L}{\partial \text{bnbias}} = \sum_i \frac{\partial L}{\partial h_{\text{post-BN}}}
$$

These gradients are then used to update $\text{bngain}$ and $\text{bnbias}$ during training.

---

### 3. **Training Both BN Parameters and Layer Weights**
The backpropagation process handles both simultaneously. Here's how:

1. **Loss Function Dependency**:
   - The loss function $L$ depends on the outputs of the network.
   - Since $\text{bngain}$ and $\text{bnbias}$ affect the outputs of the batch normalization layer, they indirectly impact the loss.
   - Backpropagation computes gradients for all parameters that influence $L$, including $\text{bngain}$, $\text{bnbias}$, and the weights/biases of layers.

2. **Independent Gradients**:
   - Gradients for $\text{bngain}$ and $\text{bnbias}$ are computed independently of the layer weights.
   - Similarly, the gradients for the layer weights (e.g., $W_1$) are computed as usual.
   - The optimization algorithm updates all parameters together.

---

### 4. **Why BN Parameters Improve Training Rates**
The optimization of $\text{bngain}$ and $\text{bnbias}$:
- Helps ensure the normalized activations are in a range suitable for learning.
- Prevents saturation of downstream activation functions (e.g., ReLU, sigmoid).
- Balances stability (from normalization) with flexibility (from learnable scale and shift).

---

### 5. **Practical Perspective**
- $\text{bngain}$ and $\text{bnbias}$ are optimized **to reduce the loss** like any other parameter, but their impact on training is indirect:
  - They improve gradient flow and learning dynamics, which accelerates convergence.
  - They adaptively scale and shift normalized activations to maintain representational power.

- Since batch normalization stabilizes training, the optimizer can focus on jointly minimizing the loss with respect to all parameters (including layer weights, biases, $\text{bngain}$, and $\text{bnbias}$) without special handling.

---

### 6. **Overall Process**
1. Forward pass:
   - Normalize activations ($\hat{h}_{\text{preact}}$).
   - Apply $\text{bngain}$ and $\text{bnbias}$.
2. Compute loss $L$.
3. Backpropagation:
   - Compute gradients for $\text{bngain}$ and $\text{bnbias}$.
   - Compute gradients for layer weights and biases.
4. Update all parameters simultaneously.

```python
# same optimization as last time
max_steps = 200000
batch_size = 32
lossi = []

for i in range(max_steps):
  
  # minibatch construct
  ix = torch.randint(0, Xtr.shape[0], (batch_size,), generator=g)
  Xb, Yb = Xtr[ix], Ytr[ix] # batch X,Y
  
  # forward pass
  emb = C[Xb] # embed the characters into vectors
  embcat = emb.view(emb.shape[0], -1) # concatenate the vectors
  # Linear layer
  hpreact = embcat @ W1 #+ b1 # hidden layer pre-activation
  # BatchNorm layer
  # -------------------------------------------------------------
  bnmeani = hpreact.mean(0, keepdim=True)
  bnstdi = hpreact.std(0, keepdim=True)
  hpreact = bngain * (hpreact - bnmeani) / bnstdi + bnbias
  # keep track of sample mean and std for final input 
  with torch.no_grad():
    bnmean_running = 0.999 * bnmean_running + 0.001 * bnmeani
    bnstd_running = 0.999 * bnstd_running + 0.001 * bnstdi
  # -------------------------------------------------------------
  # Non-linearity
  h = torch.tanh(hpreact) # hidden layer
  logits = h @ W2 + b2 # output layer
  loss = F.cross_entropy(logits, Yb) # loss function
  
  # backward pass
  for p in parameters:
    p.grad = None
  loss.backward()
  
  # update
  lr = 0.1 if i < 100000 else 0.01 # step learning rate decay
  for p in parameters:
    p.data += -lr * p.grad

  # track stats
  if i % 10000 == 0: # print every once in a while
    print(f'{i:7d}/{max_steps:7d}: {loss.item():.4f}')
  lossi.append(loss.log10().item())
  
```

# Key ideas behind a good initialization
We do not want extreme values for our weights and our biases, this will increase intial loss.
With initial high loss, we waste computational time "squishing" everything down to almost 0 before we actually start training. (this is the hockey stick shape in the previous lect's initialziation)

### Another idea is the "activeness" of the tanh function, so how much the values are "squished to 1 and -1 
We would want the tanh function to as much as possible give less extreme values
close to these saturation regions, the derivative of the activation function is close to 0.
This means the gradient signal is effectively "killed," and backpropagation fails to update the weights meaningfully.

```python
# Let's train a deeper network
# The classes we create here are the same API as nn.Module in PyTorch

class Linear:
  
  def __init__(self, fan_in, fan_out, bias=True):
    self.weight = torch.randn((fan_in, fan_out), generator=g) / fan_in**0.5
    self.bias = torch.zeros(fan_out) if bias else None
  
  def __call__(self, x):
    self.out = x @ self.weight
    if self.bias is not None:
      self.out += self.bias
    return self.out
  
  def parameters(self):
    return [self.weight] + ([] if self.bias is None else [self.bias])


class BatchNorm1d:
  
  def __init__(self, dim, eps=1e-5, momentum=0.1):
    self.eps = eps
    self.momentum = momentum
    self.training = True
    # parameters (trained with backprop)
    self.gamma = torch.ones(dim)
    self.beta = torch.zeros(dim)
    # buffers (trained with a running 'momentum update')
    self.running_mean = torch.zeros(dim)
    self.running_var = torch.ones(dim)
  
  def __call__(self, x):
    # calculate the forward pass
    if self.training:
      xmean = x.mean(0, keepdim=True) # batch mean
      xvar = x.var(0, keepdim=True) # batch variance
    else:
      xmean = self.running_mean
      xvar = self.running_var
    xhat = (x - xmean) / torch.sqrt(xvar + self.eps) # normalize to unit variance
    self.out = self.gamma * xhat + self.beta
    # update the buffers
    if self.training:
      with torch.no_grad():
        self.running_mean = (1 - self.momentum) * self.running_mean + self.momentum * xmean
        self.running_var = (1 - self.momentum) * self.running_var + self.momentum * xvar
    return self.out
  
  def parameters(self):
    return [self.gamma, self.beta]

class Tanh:
  def __call__(self, x):
    self.out = torch.tanh(x)
    return self.out
  def parameters(self):
    return []

n_embd = 10 # the dimensionality of the character embedding vectors
n_hidden = 100 # the number of neurons in the hidden layer of the MLP
g = torch.Generator().manual_seed(2147483647) # for reproducibility

C = torch.randn((vocab_size, n_embd),            generator=g)
layers = [
  Linear(n_embd * block_size, n_hidden, bias=False), BatchNorm1d(n_hidden), Tanh(),
  Linear(           n_hidden, n_hidden, bias=False), BatchNorm1d(n_hidden), Tanh(),
  Linear(           n_hidden, n_hidden, bias=False), BatchNorm1d(n_hidden), Tanh(),
  Linear(           n_hidden, n_hidden, bias=False), BatchNorm1d(n_hidden), Tanh(),
  Linear(           n_hidden, n_hidden, bias=False), BatchNorm1d(n_hidden), Tanh(),
  Linear(           n_hidden, vocab_size, bias=False), BatchNorm1d(vocab_size),
]
# layers = [
#   Linear(n_embd * block_size, n_hidden), Tanh(),
#   Linear(           n_hidden, n_hidden), Tanh(),
#   Linear(           n_hidden, n_hidden), Tanh(),
#   Linear(           n_hidden, n_hidden), Tanh(),
#   Linear(           n_hidden, n_hidden), Tanh(),
#   Linear(           n_hidden, vocab_size),
# ]

with torch.no_grad():
  # last layer: make less confident
  layers[-1].gamma *= 0.1
  #layers[-1].weight *= 0.1
  # all other layers: apply gain
  for layer in layers[:-1]:
    if isinstance(layer, Linear):
      layer.weight *= 1.0 #5/3

parameters = [C] + [p for layer in layers for p in layer.parameters()]
print(sum(p.nelement() for p in parameters)) # number of parameters in total
for p in parameters:
  p.requires_grad = True
```

```python
# same optimization as last time
max_steps = 200000
batch_size = 32
lossi = []
ud = []

for i in range(max_steps):
  
  # minibatch construct
  ix = torch.randint(0, Xtr.shape[0], (batch_size,), generator=g)
  Xb, Yb = Xtr[ix], Ytr[ix] # batch X,Y
  
  # forward pass
  emb = C[Xb] # embed the characters into vectors
  x = emb.view(emb.shape[0], -1) # concatenate the vectors
  for layer in layers:
    x = layer(x)
  loss = F.cross_entropy(x, Yb) # loss function
  
  # backward pass
  for layer in layers:
    layer.out.retain_grad() # AFTER_DEBUG: would take out retain_graph
  for p in parameters:
    p.grad = None
  loss.backward()
  
  # update
  lr = 0.1 if i < 150000 else 0.01 # step learning rate decay
  for p in parameters:
    p.data += -lr * p.grad

  # track stats
  if i % 10000 == 0: # print every once in a while
    print(f'{i:7d}/{max_steps:7d}: {loss.item():.4f}')
  lossi.append(loss.log10().item())
  with torch.no_grad():
    ud.append([((lr*p.grad).std() / p.data.std()).log10().item() for p in parameters])

  if i >= 1000:
    break # AFTER_DEBUG: would take out obviously to run full optimization
```

```python
# Linear: activation statistics of forward and backward pass

g = torch.Generator().manual_seed(2147483647)

a = torch.randn((1000,1), requires_grad=True, generator=g)          # a.grad = b.T @ c.grad
b = torch.randn((1000,1000), requires_grad=True, generator=g)       # b.grad = c.grad @ a.T
c = b @ a
loss = torch.randn(1000, generator=g) @ c
a.retain_grad()
b.retain_grad()
c.retain_grad()
loss.backward()
print('a std:', a.std().item())
print('b std:', b.std().item())
print('c std:', c.std().item())
print('-----')
print('c grad std:', c.grad.std().item())
print('a grad std:', a.grad.std().item())
print('b grad std:', b.grad.std().item())
```

```python
# Linear + BatchNorm: activation statistics of forward and backward pass

g = torch.Generator().manual_seed(2147483647)

n = 1000
# linear layer ---
inp = torch.randn(n, requires_grad=True, generator=g)
w = torch.randn((n, n), requires_grad=True, generator=g) # / n**0.5
x = w @ inp
# bn layer ---
xmean = x.mean()
xvar = x.var()
out = (x - xmean) / torch.sqrt(xvar + 1e-5)
# ----
loss = out @ torch.randn(n, generator=g)
inp.retain_grad()
x.retain_grad()
w.retain_grad()
out.retain_grad()
loss.backward()

print('inp std: ', inp.std().item())
print('w std: ', w.std().item())
print('x std: ', x.std().item())
print('out std: ', out.std().item())
print('------')
print('out grad std: ', out.grad.std().item())
print('x grad std: ', x.grad.std().item())
print('w grad std: ', w.grad.std().item())
print('inp grad std: ', inp.grad.std().item())
```
